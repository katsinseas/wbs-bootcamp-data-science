```python
import pandas as pd

tips = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/tips.csv')
```

```python
tips
```

```python
# datatypes in pandas -> series (or list/columns) / dataframe (or table)
tips.total_bill
```

```python
# Series is an object with the property of being iterable in for loops

for item in tips.total_bill:
    print(item)
    break
```

```python
# return a series with values that are false or true

series_of_boolean_values = tips.total_bill>20
```

```python
series_of_boolean_values
```

```python
series_of_boolean_values.index == tips.index
```

```python
tips[series_of_boolean_values]
# this returns the true values and the process is called Boolean Mask. Analogy of the infinite clock, when its on its showwn and when its not its hidden.
```

```python
# create a series of tips lesser or equal 3
tips_3 = tips.tip<=3
```

```python
tips_3
```

```python
tips[tips_3]
```

```python
# series can be combined, but "and" can be only used with things that are True or False, instead we use %
tips[series_of_boolean_values & tips_3] # and 
```

```python
tips[series_of_boolean_values | tips_3] # or
```

```python
tips.groupby(["sex", "smoker"])

# the message returned basically says that we still need to input operations after the "groupby"
```

```python
tips.groupby(["sex", "smoker"]).describe()
# describe returns statistics 
```

```python
tips.groupby(["total_bill", "tip"]).describe()
```

```python
tips.groupby(["time"]).sample() # this returns only one example 
```

```python
tips.groupby(["time"]).sample(n=3) # this returns 3 items for each group
```

```python
tips.groupby(["sex"]).sample()
```
